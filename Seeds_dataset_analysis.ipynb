#K Means Clustering

```python
from sklearn.cluster import KMeans
```

here x and y refers to two chosen features and not the feature and target like in other exercises

#Perimeter vs Asymmetry Coefficient

```python
x = 'perimeter'
y = 'asymmetry coefficient'
X = df[[x,y]].values
```

```python
print(X[:10])
```

```python
kmeans = KMeans(n_clusters=3, max_iter=1000).fit(X)
```

```python
clusters = kmeans.labels_
```

```python
clusters
```

```python
df['class'].values
```

combine 2 features with respective k mean clustered classes and make a new dataframe.
Reshape is needed to make clusters array from 1D to 2D

```python
cluster_df = pd.DataFrame(np.hstack((X, clusters.reshape(-1,1))), columns=[x,y,'class'])
```

```python
cluster_df.head()
```

K mean Clustered classes using perimeter and asymmetry features

```python
sns.scatterplot(x=x,y=y, hue='class', data=cluster_df)
plt.show()
```

original classes for all features plotted for perimeter vs asymmetry

```python
sns.scatterplot(x=x,y=y, hue='class', data=df)
plt.show()
```

#Principal Component Analysis

reduce the number of features of a data set, while preserving as much information as possible.

```python
from sklearn.decomposition import PCA
```

```python
X[:10]
```

reduce the features from 7 to 2

```python
pca = PCA(n_components=2)
transformed_x = pca.fit_transform(X)
```

```python
transformed_x[:10]
```

```python
plt.scatter(transformed_x[:,0], transformed_x[:,1])
plt.show()
```

Clustering the new transformed dataset using KMeans

```python
kmeans = KMeans(n_clusters=3).fit(transformed_x)
clusters = kmeans.labels_
```

```python
clusters_df = pd.DataFrame(np.hstack((transformed_x, clusters.reshape(-1,1))), columns=['pca1','pca2','class'])
```

```python
clusters_df.head()
```

transformed dataset with K Means Clustered classes displayed for pca1 vs pca2

```python
sns.scatterplot(x='pca1',y='pca2',data=clusters_df,hue='class', palette='Paired')
plt.show()
```

making a new dataframe for transformed features with original classes

```python
pca_df = pd.DataFrame(np.hstack((transformed_x, df['class'].values.reshape(-1,1))), columns=['pca1', 'pca2', 'class'])
```

```python
pca_df.head()
```

transformed dataset with original class displayed for pca1 vs pca2

```python
sns.scatterplot(x='pca1', y='pca2', data=pca_df, hue='class', palette='Paired')
plt.show()
```

By comparing the Above 2 graphs we can see that when we used K Means to cluster the transformed dataset, we get very close approximation of class to the original class for the transformed dataset
( Here we're comparing the transformed dataset with 2 features with original class vs kmeans clustered class )
